```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.get_dummies(df, dtype=int, drop_first=True)
```

```python
df.dropna(inplace=True)
```

```python
X = df.drop(columns=['Churn_Yes'])
y = df['Churn_Yes']
```

```python
y_pred = rf.predict(X_test)
```

```python
def churn_report(threshold, capacity=None):
    preds = (churn_probs >= threshold).astype(int)
    cm = confusion_matrix(y_test, preds)
    caught = cm[1,1]
    total_flagged = cm[0,1] + cm[1,1]
    false_alarms = cm[0,1]
    net_value = (cm[1,1] * 2000) - (cm[0,1] * 100)

    print(f"Threshold: {threshold}")
    print(f"total_flagged: {total_flagged}")
    print(f"caught: {caught}")
    print(f"false_alarms: {false_alarms}")
    print(f"net_value: {net_value}\n")
    
    if capacity != None:
        if total_flagged > capacity:
            print(f'⚠️ WARNING: {total_flagged} customers flagged exceeds team capacity of {capacity}!')
        else:
            print('✅ Within team capacity.')
    print(f"----------------------------------------------------------------------------------------\n\n")

for i in range(1,10):
    churn_report(i/10, 300)
```
